```python

#The fig of difficulty level of topic in SO 

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

np.random.seed(20)
test_yes = pd.read_excel('editor.xlsx')
X=test_yes['answer Percentage']
Y=test_yes['diff']

annotations=test_yes.Dominant_Topic.to_list()
plt.figure(figsize=(13,10))
sns.set(font_scale = 2)
sns.set_style("whitegrid")
sns.scatterplot('answer Percentage', 'diff', data=test_yes, hue='Dominant_Topic',s=200)
plt.ylabel("Hrs to acc answer",fontsize=18)
plt.xlabel("% No acc answer",fontsize=18)
for i, label in enumerate(annotations):

    if label =='Process statements':
        plt.text(X[i]-1, Y[i]+0.1,label,fontsize=18 )
    elif label == 'File I/O':
        plt.text(X[i]-1, Y[i]+0.1,label,fontsize=18)
    elif label =='Testing and simulation':
        plt.text(X[i]-1, Y[i]-0.2,label,fontsize=18)
    elif label == 'Bit operations':
         plt.text(X[i], Y[i]+0.1,"Bit operation",fontsize=18)
    elif label == 'Signal I/O':
         plt.text(X[i]-1, Y[i]+0.1,label,fontsize=18)
    elif label == 'Clocked processes':
         plt.text(X[i]-0.3, Y[i]+0.1,label,fontsize=18)
    elif label == 'State machines':
        plt.text(X[i]-3, Y[i]+0.1,label,fontsize=18)
    elif label == 'Memory IO':
        plt.text(X[i], Y[i]+0.1,label,fontsize=18)
    elif label == 'Data containers':
        plt.text(X[i]+0.5, Y[i]-0.1,'Data containers',fontsize=18)
    elif label == 'Type issues':
        plt.text(X[i], Y[i]+0.1,label,fontsize=18)
    elif label == 'Classes and objects':
        plt.text(X[i]-1, Y[i]+0.1,label,fontsize=18)
    elif label == 'Compiler errors':
        plt.text(X[i]-1, Y[i]+0.1,label,fontsize=18)
    elif label == 'Register operations':
        plt.text(X[i], Y[i]+0.1,label,fontsize=18)
    elif label == 'Data transfer':
        plt.text(X[i], Y[i]+0.1,label,fontsize=18)
    elif label == 'Syntax errors':
        plt.text(X[i]-3, Y[i]+0.1,label,fontsize=18)
    elif label == 'Design and implementation':
        plt.text(X[i], Y[i]+0.1,label,fontsize=18)
    else:
        print('Check label:',label)
      
    plt.legend([],[], frameon=False)
plt.show()
```
